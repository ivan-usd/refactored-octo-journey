### Use a linear regression model to predict the number of customers each airline serves from its length of flight and daily flight time per plane.

#### Linear Regression y=Customers Served x= Flight Length & Daily Flight Time

```python
df.loc[:,['FlightLength','DailyFlightTime','CustomersServed']].corr()['CustomersServed']
```

```python
y = df.CustomersServed
X = df[['FlightLength','DailyFlightTime']]
```

```python
model = sm.OLS(y,X)
results = model.fit()
```

```python
print(results.summary())
```

### Next, build another regression model to predict the total assets of an airline from the customers served by the airline

#### Linear Regression y=TotalAssets x= CustomersServed

```python
df.loc[:,['CustomersServed','TotalAssets']].corr()['TotalAssets']
```

```python
y = df.TotalAssets
X = df.CustomersServed
```

```python
model = sm.OLS(y,X)
results = model.fit()
```

```python
print(results.summary())
```
